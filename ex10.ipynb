# Exercício 10:
Mostre as volatilidades históricas dos fatores de risco Ibovespa, Jan21, USDBRL e CDS ao longo do tempo (EMWA 0.94 e GARCH - este precisa ser calibrado).

```python
from __future__ import division
import datetime
import base
import pandas as pd
import numpy as np
from dateutil.relativedelta import relativedelta
%matplotlib inline
```

Criando o método EWMA com lambda 0.94:

```python
def ewma094(dates, prices):
    ewmaLambda=0.94
    df=dates
    df['return2']=0
    nstp=len(df.index)-1
    for j in range(1,nstp+1):
        df.loc[j:,'return2']=(np.log(prices['PX_LAST'][j]/prices['PX_LAST'][j-1]))**2

    df['ewmaAbs']=0
    df.loc[1,'ewmaAbs']=df['return2'][1]
    
    for j in range(2,nstp+1):
        df.loc[j:,'ewmaAbs']=df['return2'][j]*(1-ewmaLambda) + ewmaLambda*(df['ewmaAbs'][j-1])
    
    df['ewmaPecentage']=df['ewmaAbs']*100
    return df
```

Vol histórica do Ibovespa:

```python
ewma094(base.referenceDates_df, base.ibovespa_df)
```

Vol histórica do Jan21:

```python
ewma094(base.referenceDates_df, base.di1Jan21_df)
```

Vol histórica do USDBRL:

```python
ewma094(base.referenceDates_df, base.usdbrl_df)
```

Vol histórica do CDS:

```python
ewma094(base.referenceDates_df, base.cds_df)
```
