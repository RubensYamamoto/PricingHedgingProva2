# Exercício 5:

Um outro time vendeu as opções 3.b, 3.d e 3.f; estime os resultados e perdas máximas assumindo rebalanceamento
diário.

```python
from __future__ import division
import datetime
import base
import pandas as pd
%matplotlib inline
```

#### Calculando para 3-b que está vendendo uma opção de 1m e fez o delta hedge até o final:

```python
phi=base.call
position=base.positionShort
tenor=1
startDate = datetime.datetime.strptime('06/30/2015', '%m/%d/%Y')
includeHedge=1
sb=base.calcPortfolio(phi, position, tenor, startDate, includeHedge)
sb.plot()
```

```python
pd.DataFrame([[sb[len(sb)-1], sb.min(), sb.idxmin()]], \
             columns=['PnL Final','Perda Máxima','Data Perda Máxima'])
```

#### Calculando para 3-d que está vendendo uma opção de 3m e fez o delta hedge até o final:

```python
tenor=3
sd=base.calcPortfolio(phi, position, tenor, startDate, includeHedge)
sd.plot()
```

```python
pd.DataFrame([[sd[len(sd)-1], sd.min(), sd.idxmin()]], \
             columns=['PnL Final','Perda Máxima','Data Perda Máxima'])
```
