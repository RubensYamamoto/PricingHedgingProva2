```python
from __future__ import division
import base
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

Criando o dataframe com os resultados obtidos:

```python
df=pd.DataFrame([ \
              ['3a',0.175341,-0.01039], \
              ['3b',0.213699,-0.045838], \
              ['3c',0.642843,-0.022334], \
              ['3d',0.528677,-0.048645], \
              ['3e',0.530236,-0.219451], \
              ['3f',0.425838,-0.244613], \
              ['5b',-0.213699,-0.23212], \
              ['5d',-0.528677,-1.004914], \
              ['5f',-0.425838,-0.906825], \
              ['6b',-0.23212,-0.23212], \
              ['6d',-0.818978,-0.818978], \
              ['6f',-0.906825,-0.906825], \
              ['7b',-0.306708,-0.306708], \
              ['7d',-1.471653,-1.471653], \
              ['7f',-1.236162,-1.236162], \
              ['chefe',-0.085898,-0.093518]
              ],columns=['Trader','Portfolio','Max Loss'])
df
```

Plotando o gráfico:

```python
colors = ['blue','green','red','cyan','magenta','yellow','black','darkblue', \
'darkcyan','darkgoldenrod','darkgray','darkgreen','darkkhaki','darkmagenta', \
'darkolivegreen','darkorange','darkorchid','darkred','darksalmon','darkseagreen', \
'darkslateblue','darkslategray','darkturquoise','darkviolet']


for i in range(len(df['Trader'])):
    plt.scatter(df['Max Loss'][i], df['Portfolio'][i], label=df['Trader'][i], color=colors[i])

plt.xlabel('Max Loss')
plt.ylabel('Portfolio')
plt.legend(bbox_to_anchor=(1.1, 1.1), loc='upper left', ncol=1)
plt.show()
```

O trader que obteve um melhor resultado foi o 3c, que comprou uma opção de 3m e carregou sem hedge até o final.
